__Задача 2__.

Предположим, что в некоторой популяции вероятность дожить до 60 лет равна 0.5, а вероятность дожить до 80 лет — 0.2. Какова вероятность, что случайно выбранный шестидесятилетний представитель популяции доживёт до восьмидесяти? Запишите ответ с точностью до одного знака после десятичной точки (задавать разделитель в этой задаче и следующих нужно именно в виде точки).

__Решение:__  (решение взято из [Примеры решения задач на Студопедии](https://studopedia.su/9_59944_primeri-resheniya-zadach.html "Студопедия"))


Пусть вероятность дожить до 60 лет равна p1 = 0.5, а вероятность дожить до 80 лет равна р2 = 0.2, (р2 < p1). 

Построим пространство элементарных исходов. Эксперимент заключается в выборе случайного человека и фиксировании длительности его жизни. Имеем всего три исхода: $\Omega = \{\overline{60}, 60\overline{80}, 80\}$, которые означают, что человек не дожил до 60 лет, обозначение "$60\overline{80}$" - что он дожил до 60 лет, но не дожил до 80, обозначение "80" означает, что человек дожил до 80 лет. Событие А = {человек дожил до 80 лет} влечет за собой событие В = {человек дожил до 60 лет}. По условию р(А) = р2, р(В) = p1, тогда, исходя из формулы условной вероятности $P(A|B) = \frac{P(AB)}{P(B)}$ и принимая во внимание, что вероятность наступления события B лежит полностью внутри области, в которой определено наступление события А, т.е $P(A|B) = P(A)$

```python
p1 , p2 = 0.5 , 0.2
PA = p2 / p1
print(PA)
```

__Задача 3.__

1% женщин больны раком груди. У 80% женщин, больных раком груди, маммограмма верно выявляет наличие заболевания; кроме того, она даёт ложный положительный результат (то есть, неверно показывает наличие рака) для 9.6% здоровых женщин.

У какого процента женщин, маммограмма которых дала положительный результат, есть рак груди? Запишите ответ с точностью до одного знака после десятичной точки (знак процента не нужен).

__Решение.__  ([взято здесь](http://baguzin.ru/wp/idei-bajesa-dlya-menedzherov/))

Ход рассуждений/вычислений следующий. Из 1% больных раком маммография даст 80% положительных результатов = 1%*80% = 0,8%. Из 99% здоровых женщин маммография даст 9,6% положительных результатов = 99%*9,6% = 9,504%. Итого из 10,304% (9,504% + 0,8%) с положительными результатами маммографии, только 0,8% больных, а остальные 9,504% здоровых. Таким образом, вероятность того, что при положительном результате маммографии женщина больна раком составляет 0,8%/10,304% = 7,764%. 

В нашем примере формула Байеса принимает следующий вид:

\begin{equation}
P(X_{1}|Y_{1})=\frac{p(Y_{1}|X_{1})\cdotp(X_{1})}{p(Y_{1}|X_{1})\cdotp(X_{1}) + p(Y_{1}|X_{2})\cdotp(X_{2})}
\end{equation}

Давайте еще раз проговорим «физический» смысл этой формулы. Х – случайная величина (диагноз), принимающая значения: Х1 – болен и Х2 – здоров; Y – случайная величина (результат измерения –маммографии), принимающая значения: Y1 – положительный результат и Y2 – отрицательный результат; р(Х1) – вероятность болезни до проведения маммографии (априорная вероятность), равная 1%; р(Y1|X1) – вероятность положительного результата в случае, если пациентка больна (условная вероятность, так как она должна быть задана в условиях задачи), равная 80%; р(Y1|X2) – вероятность положительного результата в случае, если пациентка здорова (также условная вероятность), равная 9,6%; р(Х2) – вероятность того, что пациентка здорова до проведения маммографии (априорная вероятность), равная 99%; р(Х1|Y1) – вероятность того, что пациентка больна, при условии положительного результата маммографии (апостериорная вероятность).

Видно, что апостериорная вероятность (то, что мы ищем) пропорциональна априорной вероятности (исходной) с несколько более сложным коэффициентом 03б. Подчеркну еще раз. На мой взгляд, это фундаментальный аспект Байесовского подхода. Измерение (Y) добавило некоторое количество информации к первоначально имевшейся (априорной), что уточнило наше знание об объекте.

__Решение мое собственное.__

Событие __(A|B)__ - 40-летняя женщина больна, при условии, что ее мамограмма положительна, то есть имеем пару взаимозависимых событий:

__А__ - событие "женщина больна";

__B__ - событие "маммограмма дала положительный результат".

__P(A)__ = 1%, поскольку в исходных условиях задачи сказано, что среди всей массы 40-летних женщин больных 1%.

Вероятность того, что маммограмма даст положительный результат, исходя из условий задачи будет равна $1\%\cdot0.8$ - это те самые 80% положительных маммограмм из 1% реально больных по условиям задачи женщин, плюс 9.6% положительных маммограмм из оставшихся 99% здоровых женщин
$\mathbf{P(B)} = 1\%\cdot0.8 + 9.6\%\cdot0.99 = 10.304\%$ .

По формуле Байеса:

\begin{equation}
P(A|B) = \frac{P(A) \cdot P(B|A)}{P(B)}
\end{equation}


Тогда имеем:

```python
from numpy.ma import round
P_AB = 80/10.304
print(round(P_AB,2))
```
