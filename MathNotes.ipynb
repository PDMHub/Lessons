
## Решение квадратных уравнений

Общий случай квадратного уравнения выглядит так:
\begin{equation}ax^{2}+bx+c=0\end{equation}

Есть квадратное уравнение, вида:
\begin{equation} \label{x1}
2x^{2}+3x+1 = 0\\
то\ есть\ a=2, b=3, c=1
\end{equation}

Решение:
1. Находим дискриминант по формуле:
\begin{equation}
D = b^{2}-4ac = 3^{2}-4\cdot2\cdot1 = 9 - 8 = 1
\end{equation}
2. Находим два корня квадратного уравнения 
\begin{equation}
x_{1} = \frac{-b + \sqrt{D}}{2\cdot{a}} = \frac{3 + \sqrt{1}}{2\cdot{2}} = -\frac{1}{2}
\end{equation}
\begin{equation}
x_{2} = \frac{-b - \sqrt{D}}{2\cdot{a}} = \frac{3 - \sqrt{1}}{2\cdot{2}} = -1
\end{equation}

## Задача нахождения собственных векторов и собственных значений матрицы.

Предположим, у нас есть матрица
\begin{equation} A=
\begin{bmatrix}
-4 & 6\\
-3 & 5
\end{bmatrix}
\end{equation}

План решения:
1. Найти корни характеристического многочлена 
\begin{equation}
\phi_{A}(\lambda) = |A-\lambda\cdot{E}| \ (правая\ части\ этого\ уравнения\ -\ это\ определитель\ некой\ разницы\ матриц\ |A-\lambda\cdot{E}| );
\end{equation}

Посольку __E__ - это единичная матрица, т.е. матрица, у которой все элементы, за исключением элементов на главной диаоганли (там единицы), равны нулю, то нам нужно по сути найти определитель такой матрицы
\begin{bmatrix}
-4 - \lambda & 6\\
-3 & 5 - \lambda
\end{bmatrix}
Определитель в конечном итоге будет равен такому уравненю:
\begin{equation}
\phi_{A}(\lambda) = \lambda^{2}-\lambda-2
\end{equation}
Решая квадратное уравнение $ \phi_{A}(\lambda) = \lambda^{2}-\lambda-2 $ , находим, что его корни $ \lambda_{1} = 2\ и \ \lambda_{2}=-1$

2. Для каждого из найденных корней многочлена найти собственный вектор по формуле $ A\cdot{v} = \lambda\cdot{v}\  ,при\ условии, что\ (v\ne0) $.
Иными словами для $\lambda_{1}=2$ нам нужно найти иксы в таком выражении:
\begin{equation}
\begin{bmatrix} -4 & 6\\ -3 & 5 \end{bmatrix} \cdot{\begin{bmatrix} x_{1}\\ x_{2} \end{bmatrix}} = 2\cdot{\begin{bmatrix} x_{1}\\ x_{2} \end{bmatrix}}
\end{equation}

\begin{equation}
\begin{bmatrix} -4\cdot{x_{1}} + 6\cdot{x_{2}}\\ -3\cdot{x_{1}} + 5\cdot{x_{2}} \end{bmatrix} = {\begin{bmatrix} 2x_{1}\\ 2x_{2} \end{bmatrix}}
\end{equation}

что равноценно системе линейных уравнений
\begin{equation}
-4x_{1} + 6x_{2} = 2x_{1}\\
-3x_{1} + 5x_{2} = 2x_{2}
\end{equation}
или, что равноценно:
\begin{equation}
-6x_{1} + 6x_{2} = 0\\
-3x_{1} + 3x_{2} = 0
\end{equation}
Эти два уравнения отличаются только множителями, поэтому фактически это одно уравнение вида $-3x_{1} + 3x_{2} = 0$.
Из него следует, что $x_{1}=x_{2}$

Поскольку достаточным условием для решения задачи является $(v\ne0)$, то, выбрав в качестве $x$ любое число, например 1, получаем, что __первый собственный вектор матрицы $A$__ равен $\begin{bmatrix} 1 \\ 1 \end{bmatrix}$.
__Второй собственный вектор__ находится аналогично первому, но уже для $\lambda_{2}=-1$
